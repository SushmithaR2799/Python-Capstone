```python
#Import Libraries

import pandas as pd
import numpy as np
```

```python
#1. Creating dataframes 

import pandas as pd

# Employee DataFrame

employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}
employee_df = pd.DataFrame(employee_data)
employee_df.to_csv('Employee.csv', index=False)

# Seniority Level DataFrame
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 2, 3, 2, 3]
}
seniority_df = pd.DataFrame(seniority_data)
seniority_df.to_csv('Seniority.csv', index=False)

# Project DataFrame
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}
project_df = pd.DataFrame(project_data)
project_df.to_csv('Project.csv', index=False)
```

```python
#2. Replacing missing values with running average with "For" loop

project_df = pd.read_csv('Project.csv')

# Calculate running average for non-null values
running_sum = 0
count = 0
for i in range(len(project_df)):
    if pd.isnull(project_df.loc[i, 'Cost']):
        project_df.loc[i, 'Cost'] = running_sum / count if count != 0 else 0
    else:
        running_sum += project_df.loc[i, 'Cost']
        count += 1
        
#Saving the dataframe to CSV file. 
project_df.to_csv('Project.csv', index=False)

#Printing the output
project_df
```

```python
#3. Splitting the "name" column to "First Name" and "Last Name"

import pandas as pd

# Loading the employee DataFrame
employee_df = pd.read_csv('Employee.csv')

# Splitting the 'Name' column into 'First Name' and 'Last Name'
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', n=1, expand=True)

# Dropping the old 'Name' column
employee_df.drop(columns=['Name'], inplace=True)

#Saving the dataframe to CSV file. 
employee_df.to_csv('Employee.csv', index=False)

#Printing the output
employee_df
```

```python
#4. Joining all Dataframes:

employee_df = pd.read_csv('Employee.csv')
seniority_df = pd.read_csv('Seniority.csv')
project_df = pd.read_csv('Project.csv')

# Merging the DataFrames
final_df = pd.merge(employee_df, seniority_df, on='ID')
final_df = pd.merge(final_df, project_df, on='ID')

#Saving the dataframe to CSV file. 
final_df.to_csv('Final.csv', index=False)

#Printing the output
final_df
```

```python
#5. Adding new bonus column:

final_df = pd.read_csv('Final.csv')

# Adding bonus column
final_df['Bonus'] = final_df.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)

final_df.to_csv('Final.csv', index=False)

#Printing the output
final_df
```

```python
#6. Demoting designation levels:

final_df = pd.read_csv('Final.csv')

# Demote designation level for failed projects
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] += 1

# Delete employees whose designation level is above 4
final_df = final_df[final_df['Designation Level'] <= 4]

final_df.to_csv('Final.csv', index=False)

#Printing the output
final_df
```

```python
#7. Adding honorifics and dropping gender column. 

final_df = pd.read_csv('Final.csv')

# Adding titles
final_df['First Name'] = final_df.apply(lambda row: 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name'], axis=1)

# Dropping gender column
final_df.drop(columns=['Gender'], inplace=True)

final_df.to_csv('Final.csv', index=False)

#Printing the output
print(final_df)
```

```python
#8. Promoting Designation level by age:

final_df = pd.read_csv('Final.csv')

# Promote designation level by age
final_df['Designation Level'] = final_df.apply(lambda row: min(4, row['Designation Level'] + 1) if row['Age'] > 29 else row['Designation Level'], axis=1)

final_df.to_csv('Final.csv', index=False)

#Printing the output
final_df
```

```python
#9. Adding total project cost for employees:

final_df = pd.read_csv('Final.csv')

# Calculate total project cost for each employee
total_project_cost = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()
total_project_cost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

total_project_cost.to_csv('TotalProjCost.csv', index=False)

#Printing the output
final_df
```

```python
#10. Printing employee details with specific city names:

final_df = pd.read_csv('Final.csv')

# Filtering employees based on city names containing 'o'
filtered_employees = final_df[final_df['City'].str.contains('o', case=False)]

#Printing the output
print(filtered_employees)
```
